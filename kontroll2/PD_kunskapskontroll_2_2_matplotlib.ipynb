### Plotting in Matplotlib can be done in either of two ways, which ones? Which way is the recommended approach?

```python
It can be done explicitly, i.e. object oriented, or implicitly which gives less control.
The explicit way is recommended.
```

### Explain shortly what a figure, axes, axis and an artist is in Matplotlib.

```python
A figure is an object which contains every other object (artists) that occurs in a diagram.
An axes is a region (artist) where data can be plotted on a diagram. One or more of them are contained in a figure.
An axis is an object (artist) that sets scale, limits, ticks and ticklabels in a diagram. One per dimension.
All visible graphical objects that occur on a canvas of a figure are artists, like text, lines, axes, axis etc. Even figures.
```

### When plotting in Matplotlib, what is the expected input data type?

```python
Objects of type numpy array or objects that can be passed to numpy.asarray.
```

### Create a plot of the function y = x^2 [from -4 to 4, hint use the np.linspace function] both in the object-oriented approach and the pyplot approach. Your plot should have a title and axis-labels.

```python
import numpy as np
y = np.linspace(-4, 4, 9)

#explicit approach
fig, ax = plt.subplots()
ax.set_title("y=x^2")
ax.plot(y, y**2)
ax.set_ylabel("y axis!")
ax.set_xlabel("x axis!")
plt.show()

#implicit approach
plt.title("y=x^2")
plt.plot(y, y**2)
plt.ylabel("y axis")
plt.xlabel("x axis")
plt.show()
```

### Create a figure containing 2  subplots where the first is a scatter plot and the second is a bar plot. You have the data below. 

```python
# Data for scatter plot
np.random.seed(15)
random_data_x = np.random.randn(1000)
random_data_y = np.random.randn(1000)
x = np.linspace(-2, 2, 100)
y = x**2

# Data for bar plot
fruit_data = {'grapes': 22, 'apple': 8, 'orange': 15, 'lemon': 20, 'lime': 25}
names = list(fruit_data.keys())
values = list(fruit_data.values())
```

```python
# scatter plot
fig, ax = plt.subplots(1, 2, figsize=(9, 3))
ax[0].scatter(random_data_x, random_data_y, alpha=0.5)

# bar plot
ax[1].bar(x=names, height=values)

plt.show()
```
